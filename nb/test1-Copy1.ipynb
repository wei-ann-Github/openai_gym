```python
import matplotlib.pyplot as plt
from time import sleep

import gymnasium as gym
import numpy as np
from pettingzoo.classic import connect_four_v3

env = connect_four_v3.env(render_mode='human')
```

```python
obs = env.reset()
print(f'obs: {obs}')

for agent in env.agent_iter():
    
    observation, reward, done, truncated, info = env.last()
    if done:
        print(observation)
        break
    
    print(agent)
    
    # pick an action
    valid_moves = np.where(observation['action_mask'])[0]
    action = np.random.choice(valid_moves, 1)[0]
#     action = env.action_space(agent=agent).sample()  # pick a random action. But this ignores valid moves.
#     action = policy(observation, agent)  # pick an action based on a policy
    
    
    print(f'{agent} action: {action} (put in column {action + 1})')
    
    env.step(action)
    env.render()
    sleep(0.5)

print(observation['action_mask'])
# env.close()  # to close the render properly.

draw = True
for player, outcome in env.rewards.items():
    if outcome > 0:
        draw = False
        print(f'{player} won')
        
if draw:
    print(f'game ended in a draw')
    
env.render()
```
